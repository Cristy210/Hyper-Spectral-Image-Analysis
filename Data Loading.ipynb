```python
import numpy as np
import matplotlib.pyplot as plt
import os
import rasterio
```

```python
main_dir = '/Users/suhasmathey/Hyper_Spectral Imaging/Hyper-Spectral-Image-Analysis/Onera Satellite Change Detection dataset - Images'
```

```python
train_cities1 = ["abudhabi", "mumbai", "aguasclaras", "beihai", "beirut", "bercy", "bordeaux", "cupertino", "hongkong", "nantes", "paris", "pisa", "rennes", "saclay_e"]
```

```python
Train_T1 = {}
```

```python
for city in train_cities1:
    city_path = os.path.join(main_dir, city, 'imgs_1_rect')
    matrices = []

    for file_name in os.listdir(city_path):
        if file_name.endswith('.tif'):
            tif_path = os.path.join(city_path, file_name)

            with rasterio.open(tif_path) as src:
                matrix = src.read(1)
                matrices.append(matrix)
    city_tensor = np.stack(matrices, axis=0)
    Train_T1[city] = city_tensor
```

```python
plt.figure(figsize=(6, 6))
plt.imshow(Train_T1['mumbai'][3], cmap='gray')
plt.show()
```
